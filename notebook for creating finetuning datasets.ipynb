```python
# getting all sorts of https errors - this was an attempt to fix this - probably only need verify = False as done in requests call
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.ssl_ import create_urllib3_context

class CustomHTTPAdapter(HTTPAdapter):
    def init_poolmanager(self, *args, **kwargs):
        context = create_urllib3_context()
        context.set_ciphers('DEFAULT@SECLEVEL=1')
        kwargs['ssl_context'] = context
        return super(CustomHTTPAdapter, self).init_poolmanager(*args, **kwargs)

session = requests.Session()
session.mount('https://', CustomHTTPAdapter())
```

```python
data_dirs = ['classics-data', 'euripides-data']

contains = ['Provided by The Internet Classics Archive.', 'See bottom for copyright. Available online at', 'Translated by ', 'THE END', 'Copyright statement:', '----------------------------------------------------------------------', '<html><body><pre>', '</pre></body></html>']

# iterate through contains and for each see if in file
def check_file(file):
	with open(file) as f:
		text = f.read()
		for c in contains:
			if c in text:
				pass
				#print(f'\t{c} found')
			else:
				print(f'\t{c} not found')

for data_dir in data_dirs:
	for file in glob.glob(f'{data_dir}/*.txt'):
		print(file)
		check_file(file)
```

```python
# create cleaned dir for each data_dir and add to cleaned_dirs
cleaned_dirs = []
for data_dir in data_dirs:
	cleaned_dir = data_dir.replace('-data', '-cleaned')
	cleaned_dirs.append(cleaned_dir)
	if not os.path.exists(cleaned_dir):
		os.makedirs(cleaned_dir)
```

```python
contains = ['Provided by The Internet Classics Archive.', 'See bottom for copyright. Available online at', 'Translated by ', 'THE END', 'Copyright statement:', '----------------------------------------------------------------------']

# iterate through all files in data_dir and remove contains
def clean_file(file, cleaned_file):
	with open(file) as f:
		text = f.read()
		# remove all text before and including '<html><body><pre>'
		# remove all text after and including  '</pre></body></html>'
		text = text.split('<html><body><pre>')[-1].split('</pre></body></html>')[0]
		# remove everthing after and including 'The Internet Classics Archive by '
		text = text.split('The Internet Classics Archive by ')[0]
		# if there is a line with items in contains remove that line
		lines = text.split('\n')
		lines = [line for line in lines if not any(c in line for c in contains)]
		# join lines back together
		text = '\n'.join(lines)
		# remove any urls
		text = re.sub(r'https?://\S+', '', text)
		# convert to unix line endings
		text = text.replace('\r\n', '\n')
		## two newlines are ok, but more than two newlines should be reduced to two newlines
		text = re.sub(r'\n{3,}', '\n\n', text)
		# save
		with open(cleaned_file, 'w') as f:
			f.write(text.strip())

for data_dir, cleaned_dir in zip(data_dirs, cleaned_dirs):
	for file in glob.glob(f'{data_dir}/*.txt'):
		filename = file.split('\\')[-1]
		cleaned_file = os.path.join(cleaned_dir, filename)
		print(file, ' > ' , cleaned_file)
		clean_file(file, cleaned_file)
```

```python
# iterate through all files and show me all lines less than 20 characters
for cleaned_dir in cleaned_dirs:
	for file in glob.glob(f'{cleaned_dir}/*.txt'):
		with open(file) as f:
			lines = f.readlines()
			for line in lines:
				if len(line) < 20 and line.strip() != '':
					print(file, line.strip())
```

```python
# create new processed_dir for each cleaned_dir and add to processed_dirs
processed_dirs = []
for cleaned_dir in cleaned_dirs:
	processed_dir = cleaned_dir.replace('-cleaned', '-chunked')
	processed_dirs.append(processed_dir)
	if not os.path.exists(processed_dir):
		os.makedirs(processed_dir)
```

```python
# iterate through all files in cleaned_dir and split into chunks based on paragraphs
# ignore BOOK I, BOOK II, etc. and ACT I, ACT II, etc.
# group Dramatis Personae with the next paragraph
# first line is title
# second line is author
# add to a csv with fields: filename, title, author, chunk_number, chunk_text

def process_file(file, processed_file):
	# chunk into paragraphs (two newlines)
	with open(file) as f:
		text = f.read()
		chunks = text.split('\n\n')
		# remove any chunks that are BOOK I, BOOK II, etc. and ACT I, ACT II, etc.
		chunks = [chunk for chunk in chunks if not re.match(r'BOOK [IVXLCDM]+', chunk) and not re.match(r'ACT [IVXLCDM]+', chunk)]
		# group Dramatis Personae with the next paragraph
		new_chunks = []
		i = 0
		while i < len(chunks):
			chunk = chunks[i]
			if 'DRAMATIS PERSONAE' in chunk.upper():
				new_chunk = chunk + '\n\n' + chunks[i+1]
				new_chunks.append(new_chunk)
				i += 2
			else:
				new_chunks.append(chunk)
				i += 1
		# first line in first chunk is title
		# second line in first chunk is author
		title = new_chunks[0].split('\n')[0]
		author = new_chunks[0].split('\n')[1]
		# remove By from author
		author = author.replace('By ', '')
		# remove title and author
		new_chunks = new_chunks[1:]
		# add to a csv with fields: filename, title, author, chunk_number, chunk_text
		with open(processed_file, 'w', newline='', encoding='utf8') as f:
			writer = csv.writer(f)
			writer.writerow(['filename', 'title', 'author', 'chunk_number', 'chunk_text'])
			for i, chunk in enumerate(new_chunks):
				writer.writerow([file.split('\\')[-1], title, author, i, chunk])

for cleaned_dir, processed_dir in zip(cleaned_dirs, processed_dirs):
	for file in glob.glob(f'{cleaned_dir}/*.txt'):
		filename = file.split('\\')[-1]
		processed_file = os.path.join(processed_dir, filename).replace('.txt', '.csv')
		print(file, ' > ' , processed_file)
		process_file(file, processed_file)

```

```python
combined_file = 'classics-combined.csv'
```

```python
# iterate through chunked files and combine into one csv
# add a field for text_length

with open(combined_file, 'w', newline='', encoding='utf8') as f:
	writer = csv.writer(f)
	writer.writerow(['filename', 'title', 'author', 'chunk_number', 'chunk_text', 'text_length'])
	for processed_dir in processed_dirs:
		for file in glob.glob(f'{processed_dir}/*.csv'):
			with open(file, newline='', encoding='utf8') as f:
				reader = csv.reader(f)
				next(reader)
				for row in reader:
					text_length = len(row[4])
					writer.writerow(row + [text_length])
```
